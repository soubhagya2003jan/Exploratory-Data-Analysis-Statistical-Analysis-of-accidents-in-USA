## Exploratory Data Analysis (EDA)

Exploratory Data Analysis (EDA) is a critical step in the data analysis process that involves summarizing and visualizing the main characteristics of a dataset. The primary objectives of EDA include:

- **Understanding Data Distribution**: Assessing the distribution of numerical and categorical variables.
- **Identifying Patterns and Trends**: Discovering relationships and trends between variables.
- **Detecting Outliers**: Identifying anomalies that may affect analysis.
- **Assessing Data Quality**: Evaluating the dataset for missing values, duplicates, and inconsistencies.
- **Visualizations**: Using graphical techniques (like histograms, scatter plots, and box plots) to facilitate pattern recognition.

## Statistical Analysis

Statistical Analysis is the process of collecting, analyzing, interpreting, and presenting data to uncover patterns, trends, and relationships. It involves applying statistical techniques to draw conclusions and make informed decisions based on data.

Statistical Analysis is essential in various fields, including business, healthcare, social sciences, and engineering, as it provides a foundation for data-driven decision-making.

```python
plt.figure(figsize=(10, 8))
ax = sns.barplot(x=cities_by_accident[:20].index, y=cities_by_accident[:20].values, color='skyblue')

for index, value in enumerate(cities_by_accident[:20].values):
    plt.text(index, value, f'{int(value)}', ha='center', va='bottom')  # Display the value above the bar

plt.title('Accidents Count by Top 20 Cities')
plt.xlabel('Cities')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

Most Frequent accidents in the daytime

```python
data['Start_Time'] = pd.to_datetime(data['Start_Time'], format='%Y-%m-%d %H:%M:%S', errors='coerce')

data = data.dropna(subset=['Start_Time'])

data['Hour'] = data['Start_Time'].dt.hour

accidents_by_hour = data['Hour'].value_counts().sort_index()

plt.figure(figsize=(12, 8))
ax = sns.barplot(x=accidents_by_hour.index, y=accidents_by_hour.values, color='red')
plt.title('Number of Accidents by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Accidents')

for index, value in enumerate(accidents_by_hour.values):
    plt.text(index, value, str(value), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
data['Start_Time'] = pd.to_datetime(data['Start_Time'], errors='coerce')
data['Day_of_Week'] = data['Start_Time'].dt.day_name()
accidents_by_day = data['Day_of_Week'].value_counts().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'], fill_value=0)

plt.figure(figsize=(10, 6))
sns.barplot(x=accidents_by_day.index, y=accidents_by_day.values, palette='bright')
plt.title('Number of Accidents by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Accidents')

for index, value in enumerate(accidents_by_day.values):
    plt.text(index, value + 2, str(value), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

Accidents by year(Trend)

```python
data['Year'] = pd.to_datetime(data['Start_Time']).dt.year

accidents_per_year = data['Year'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
sns.lineplot(x=accidents_per_year.index, y=accidents_per_year.values, marker='o')

plt.title('Number of Accidents per Year')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.xticks(accidents_per_year.index)  
plt.grid()
for i, value in enumerate(accidents_per_year.values):
    plt.text(accidents_per_year.index[i], value, str(value), 
             ha='center', va='bottom')
plt.show()
```

Visualizing the accidents by wind data

```python
data['Wind_Speed(mph)'] = pd.to_numeric(data['Wind_Speed(mph)'], errors='coerce')
data = data.dropna(subset=['Wind_Speed(mph)'])

bins = [0, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50]
labels = ['0-5 mph', '6-10 mph', '11-15 mph', '16-20 mph', '21-25 mph', '26-30 mph', '31-35 mph', '36-40 mph', '41-45 mph', '46+ mph']

data['Wind_Speed_Category'] = pd.cut(data['Wind_Speed(mph)'], bins=bins, labels=labels, right=False)

accidents_by_wind_speed = data['Wind_Speed_Category'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
sns.barplot(x=accidents_by_wind_speed.index, y=accidents_by_wind_speed.values, palette='viridis')

plt.title('Number of Accidents by Wind Speed')
plt.xlabel('Wind Speed (mph)')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.grid()

plt.tight_layout()
plt.show()
```

```python
data.loc[:, 'Wind_Direction'] = data['Wind_Direction'].astype(str)
data = data[data['Wind_Direction'].notnull()]

accidents_by_wind_direction = data['Wind_Direction'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=accidents_by_wind_direction.index, y=accidents_by_wind_direction.values)

plt.title('Number of Accidents by Wind Direction')
plt.xlabel('Wind Direction')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)
plt.grid()

plt.tight_layout()
plt.show()
```
